```julia
using LinearAlgebra
# Problem F(x) = 0
# lam = 3.82843
#lam = 4

lam = 2

tol = 1e-12
p = Inf

F((x, y, z)) = [x - lam*z*(1-z)
    y - lam*x*(1-x)
    z - lam*y*(1-y)]

DF((x, y, z)) = [1 0 -lam*(1-2*z);
    -lam*(1-2*x) 1 0;
    0 -lam*(1-2*y) 1]

# Newton iteration
X0 = [1; -1; 1]
X = X0

num_itr = 0
FX = F(X)
println("Before iteration: $(norm(FX,p))")

while num_itr ≤ 100
    dx = DF(X)\FX
    X = X - dx
    num_itr += 1
    FX = F(X)
    println("After $(num_itr) iteration: $(norm(FX,p))")
    if norm(FX,p) < tol
        break
    end
end
X_bar = X

println(X_bar)
```

```julia
# A posteriori validation
A = inv(DF(X_bar))
@show Y0 = norm(A*F(X_bar),p)
@show Z0 = opnorm(Matrix{Float64}(I,size(A)) - A*DF(X_bar),p)
Z1 = 0
@show Z2 = 2*lam*opnorm(A,p)

# p(r) = Z2*rˆ2 - (1-Z1-Z0)*r + Y0
a = Z2
b = -(1-Z1-Z0)
c = Y0

if b^2-4*a*c < 0
    println("error: cannot find root of radii polynomial")
else
    r_min = (-b - sqrt(b^2-4*a*c))/2/a
    r_max = (-b + sqrt(b^2-4*a*c))/2/a
    r_theta = r_min
    println("The solution is included in the interval vector")
    @show X_bar
    println("with the radius $(r_theta).")
end
```
